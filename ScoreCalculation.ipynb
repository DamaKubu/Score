# atrocity implementation:



|S| reiškia absoliutų S dydį, kadangi S yra aibė tikriausiai turi omeny kardinalumą, t.y. kiek elementų aibėje.

O per visą sigmą einantis || čj reiškia tsg sveikają skaičiaus dalį.



```python
P = 1_000_000
D = 99 / 100

T = {0: 'red', 1: 'blue'}  #komanda 0-'red', 1-'blue'

S = ['pyrago serija', 'tango serija']

C = {}
C['pyrago serija'] = [
    'išmaišyt tešla', 'įdėt į orkaite', 'nušaut kaimyną', 'valgyt pyragą'
]
C['tango serija'] = ['cabaceo', 'šokis', 'parsivedimas ir atsisveikinmas']

#print([[R[1][c] for c in C[s]] for s in S])
#print([[c for c in C[s]] for s in S])

# cia gal loookup table kazkoki
R = [[]]  #[['red'], ['blue']]
#klausimas ar yra verta atskirt uzduoteles ar tsg sudet viska i viena didele serija komandai
#nu aisku su pandas biski graziau, bet for the sake of dumb illustration
R[0] = {
    'išmaišyt tešla': 0,
    'įdėt į orkaite': 0,
    'nušaut kaimyną': 1,
    'valgyt pyragą': 0,
    'cabaceo': 1,
    'šokis': 1,
    'parsivedimas ir atsisveikinmas': 1
}
R.append({
    'išmaišyt tešla': 1,
    'įdėt į orkaite': 1,
    'nušaut kaimyną': 0,
    'valgyt pyragą': 1,
    'cabaceo': 0,
    'šokis': 0,
    'parsivedimas ir atsisveikinmas': 0
})
```

```python
def p(x):
    pp = [[
        int(P * D**R[x][c] *
            (len(S) * len(C[s]) * np.sum([D**R[t][c] for t in T]))**-1)
        for c in C[s]
    ] for s in S]
    return sum([sum(pp[i]) for i in range(len(S))])
```
